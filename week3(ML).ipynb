```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
data=load_iris()
X=data.data
y=data.target
target_names=data.target_names
print("Original Data Shape:",X.shape)
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
print("Scaled Data:",X_scaled)
minmax=MinMaxScaler()
X_normal=minmax.fit_transform(X)
print("Normalized Data:",X_normal)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
print(X_pca)
```

```python
print("\nExplained Variance Ratio:")
for i,var in enumerate(pca.explained_variance_ratio_):
  print(f"Principal Component {i+1}: {var:.4f}")
print("\nCumulative Explained Variance:",np.cumsum(pca.explained_variance_ratio_))
```

```python
plt.figure()
for i,target in enumerate(target_names):
  plt.scatter(X_pca[y==i,0],X_pca[y==i,1],label=target)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.title("PCA (2D Projection)")
plt.show()
```

```python
pca_full=PCA()
pca_full.fit_transform(X_scaled)
plt.figure()
plt.plot(np.cumsum(pca_full.explained_variance_ratio_),marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.grid()
plt.show()
```
